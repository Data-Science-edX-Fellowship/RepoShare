### Qué hace autoML

AutoML trata de encontrar convinaciones de modelos y de hisperparametros que maximicen la precision de los modelos.Para ello:

- AutoML elige una estrategia para preprocesar los datos: cómo tratar los datos desequilibrados, cómo llenar los valores faltantes, eliminar, reemplazar o mantener valores atípicos, cómo codificar categorías y columnas de categorías múltiples, cómo prevenir errores de memoria...
- AutoML genera nuevas variables y selecciona las significativas
- AutoML selecciona el modelo (modelos lineales, K-vecinos más cercanos, aumento de gradiente, redes neuronales...)
- AutoML ajusta los hiperparámetros del modelo elegido 
- AutoML crea un conjunto estable de modelos para aumentar la puntuación si es posible


**Sistemas AutoML**

A lo largo de los últimos años se han desarrollado varios paquetes de aprendizaje automático automático. Aunque hay más paquetes que los que se enumeran a continuación, algunos de los más conocidos son:

- **AutoWEKA** es un enfoque para la selección simultánea de un algoritmo de aprendizaje automático y sus hiperparámetros; combinado con el paquete WEKA, produce automáticamente buenos modelos para una amplia variedad de conjuntos de datos.

- **Auto-sklearn** basado en la biblioteca de Python scikit-learn, que es un reemplazo directo para clasificadores y regresores regulares de scikit-learn.

- **TPOT** es un asistente de ciencia de datos que optimiza las canalizaciones de aprendizaje automático mediante programación genética.

- **H2O AutoML** proporciona selección automatizada de modelos y se asemeja a la plataforma de aprendizaje automático y análisis de datos H2O.

- **Pycaret** es una de las librerias de autoML más nueva y completas, es de codigo abierto

# Auto-Skelearn

Auto-sklearn permite automatizar el aprendizaje supervisado convencional. Esta construido en torno a la biblioteca de aprendizaje automático scikit-learn.
Auto-sklearn busca automáticamente el algoritmo de aprendizaje adecuado para un nuevo conjunto de datos y optimiza sus hiperparámetros. Por lo tanto, permite automatizar las tareas más tediosas del proceso de Data Science, para así poder centrarnos en lo que más valor aporta.

# Comandos básicos auto_Sklearn

Auto-Sklearn posee una sintaxis y modo de trabajo muy similar a la de scikyt-learn. Por ejemplo, en el caso de la clasificación de numeros con el dataset de MNIST, la estructura básica y los comandos serían los siguientes. 

En primer lugar, tenemos que ejecutar algunas importaciones:
1. <code>autosklearn.classification </code> que usaremos para cargar el clasificador más adelante
2. <code>sklearn.model_selection </code>para la selección de modelos
3. <code>sklearn.datasets </code>para cargar en el conjunto de datos
4. <code>import sklearn.metrics </code> para medir el rendimiento del modelo

```python
import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
```

Como de costumbre, cargamos en el conjunto de datos y lo dividimos en un conjunto de entrenamiento y prueba. A continuación, importamos <code>AutoSklearnClassifier</code> de <code>autosklearn.classification</code>. Una vez hecho esto, ajustamos el clasificador a nuestro conjunto de datos, hacemos predicciones y comprobamos la precisión

```python
X, y = sklearn.datasets.load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=1)
```

```python
automl = autosklearn.classification.AutoSklearnClassifier()
automl.fit(X_train, y_train)
```

```python
y_hat = automl.predict(X_test)
print("Score of Accuracy", sklearn.metrics.accuracy_score(y_test, y_hat))
```
