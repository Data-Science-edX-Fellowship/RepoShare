### Task 3.1: Calculating the first derivative of the curve

```python
df.loc["China"].diff().plot()
plt.show()
```

### Task 3.2: Find maximum infection rate for China

```python
df.loc["China"].diff().max()
```

```python
df.loc["Italy"].diff().max()
```

```python
df.loc["Spain"].diff().max()
```

### Task 3.3: Find maximum infection rate for all of the countries. 

```python
countries = list(df.index)
max_infection_rates = []
for country in countries:
    max_infection_rates.append(df.loc[country].diff().max())
df['max infection rate'] = max_infection_rates
```

```python
df.head()
```

### Task 3.4: Create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(df['max infection rate'])
```

```python
corona_data
```
